```python
obj_gender_df = df.select_dtypes(include=['object']).copy()
obj_gender_df.head()
```

```python
# I will start by using OneHotEncoder on the 'gender' column. Employing OneHotEncoder means 
# we are adding new columns for each unique value we need to encode within that column
# since 'gender' only has F or M as unique values that need to be encoded, it adds only 2 new columns

from sklearn.preprocessing import OneHotEncoder
oe_gender = OneHotEncoder()
oe_gender_results = oe_gender.fit_transform(obj_gender_df[["gender"]])
pd.DataFrame(oe_gender_results.toarray(), columns=oe_gender.categories_).head()
```

```python
obj_gender_df = obj_gender_df.join(pd.DataFrame(oe_gender_results.toarray(), columns=oe_gender.categories_))
```

```python
obj_gender_df.head(5)
```

```python
# Here I will try something more complex to encode, such as the state
df.state.value_counts()
```

```python
obj_state_df = df.select_dtypes(include=['object']).copy()
obj_state_df.head()
```

```python
oe_state = OneHotEncoder()
oe_state_results = oe_state.fit_transform(obj_state_df[["state"]])
pd.DataFrame(oe_state_results.toarray(), columns=oe_state.categories_).head()
```

```python
obj_state_df = obj_state_df.join(pd.DataFrame(oe_state_results.toarray(), columns=oe_state.categories_))
```
